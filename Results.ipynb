```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read(name):
    buff = None
    with open(name, 'r') as f:
        buff = f.read()

    return json.loads(buff)
```

```python
def set_axis(ax):
    ax.set_axisbelow(True)
    ax.grid()
    ax.legend()
```

```python
final = read('models/model_100_100_final.json')
final.keys()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (9, 4))

ax1.plot(np.arange(1, 100), final['class_out_loss'][1:], label='Crossentropy')
ax1.set_xlabel('Epochs')
set_axis(ax1)

ax2.plot(np.arange(1, 100), final['box_out_loss'][1:], label='MSE')
ax2.set_xlabel('Epochs')
set_axis(ax2)

plt.savefig("images/crossentropy_mse.pdf", bbox_inches = 'tight', pad_inches = 0)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (9, 4))

ax1.plot(final['class_out_accuracy'], label='Accuracy')
ax1.set_xlabel('Epochs')
set_axis(ax1)

ax2.plot(final['box_out_iou'], label='IoU')
ax2.set_xlabel('Epochs')
set_axis(ax2)

plt.savefig("images/accuracy_iou.pdf", bbox_inches = 'tight', pad_inches = 0)
```
